```python
import numpy as np
import sympy as sp #Algebra Rechner sowie Wolfram Alpha
import pylab as pl #plotten
pl.ion()
```

```python
x, E, t = sp.symbols('x E t') #AlgebraAusdrücke
```

```python
phi =  sp.exp(-(x-4*t)**2/(4*E*(t+1))) + sp.exp(-(x-4*t-2*np.pi)**2/(4*E*(t+1))) #DGL-Lösung
```

```python
phiprime = phi.diff(x) #Ableitung
```

```python
u =  -2*E*(phiprime/phi)+4 #Burgersgleichung
```

```python
from sympy.utilities.lambdify import lambdify
ufunc =  lambdify ((t, x, E), u) # u konvertieren in lambda funktionen mit den den Variablen t,x, Elasti

```

```python
nx = 101
nt = 100
dx = 2*np.pi/(nx-1)
E = 0.4
dt = dx*E
T = nt*dt

```

```python
grid =  np.linspace(0, 2*np.pi, nx)
un = np.empty(nx)
t = 0
```

```python
u = np.asarray([ufunc(t, x, E) for x in grid])
```

```python
pl.figure(figsize=(11,7), dpi=100)
pl.plot(grid, u, marker='o', lw=2)
pl.xlim([0,2*np.pi])
pl.ylim([0,10])
pl.xlabel('X')
pl.ylabel('Velocity')
pl.title('1D Burgers Equation - Initial condition')
```
